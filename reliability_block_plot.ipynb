```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
lambda_ = 0.1  # Decay rate (example value)
t = np.linspace(0, 50, 500)  # Time from 0 to 50

# Mortality rate function for the essential module
def mortality_rate(t, n, lambda_):
    e_lambda_t = np.exp(-lambda_ * t)
    numerator = n * lambda_ * e_lambda_t * (1 - e_lambda_t)**(n - 1)
    denominator = 1 - (1 - e_lambda_t)**n
    return numerator / denominator

# Calculate mortality rates for n=1 and n=2
mu_m1 = mortality_rate(t, 1, lambda_)
mu_m2 = mortality_rate(t, 2, lambda_)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, mu_m1, label='n=1')
plt.plot(t, mu_m2, label='n=2')
plt.title('Mortality Rate of the Essential Module, $\mu_m(t)$')
plt.xlabel('Time, $t$')
plt.ylabel('Mortality Rate, $\mu_m(t)$')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate mortality rates for n=1, 2, 3, and 4
mu_m3 = mortality_rate(t, 3, lambda_)
mu_m4 = mortality_rate(t, 4, lambda_)

# Plotting expanded to n=1, 2, 3, and 4
plt.figure(figsize=(12, 8))
plt.plot(t, mu_m1, label='n=1')
plt.plot(t, mu_m2, label='n=2')
plt.plot(t, mu_m3, label='n=3')
plt.plot(t, mu_m4, label='n=4')
plt.title('Mortality Rate of the Essential Module, $\mu_m(t)$')
plt.xlabel('Time, $t$')
plt.ylabel('Mortality Rate, $\mu_m(t)$')
plt.legend()
plt.grid(True)
plt.show()
```
